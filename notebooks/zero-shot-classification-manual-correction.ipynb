```python
import pandas as pd
data = pd.read_csv('/kaggle/input/data-cleaned/cleaned_data.csv')
data.head()
```

```python

labels = ["سياسة","اجتماعي" ,"اقتصاد", "رياضة", "تكنولوجيا", "ثقافة", "صحة"]
```

```python
import transformers
transformers.__version__
```

```python
from transformers import pipeline
```

```python
classifier = pipeline("zero-shot-classification", model="MoritzLaurer/mDeBERTa-v3-base-mnli-xnli")
```

```python
data["label_model"] = data['title']
```

```python
for inx , text in enumerate(data['clean_combined_text']): # 222 datarow
    result = classifier(text , labels)
    data['label_model'][inx] = result['labels'][0]
    print(f"the index {inx}")
    print(f"text : {text}")
    print("pred label ", result['labels'][0])
    print("the scores ", result["scores"])
```

```python
data.head()
```

```python
data.to_csv('cleaned_data_with_labels.csv', index=False , encoding= "utf-8-sig")
```
